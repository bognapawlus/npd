# Main project 

We have given tables with data for some countries in some years

* `population` – table with given population 
* `gpd` – table with given gpd 
* `co2_emmisions` – table with given co2_emissions, with different format than the two tables above (see preview of tables below). Moreover this table has some types in country names.

We have following tasks:
* **task1:** Displaying the table of 5 countries with greatest value of gdp / population for each year
* **task2:** Displaying the table of 5 countries with greatest co2 emissions per person for each year

## Instalation:
1. Clone the repo:
```sh
 $ git clone https://github.com/bognapawlus/npd/tree/master/main_project
```
2. Run by
```sh
 $ python3 projekt.py [-h] [-g | -c] [-s STARTYEAR] [-e ENDYEAR] population.csv gdp.csv co2_emission.csv
```
description of options:
*  `-h` shows info about usage
*  `-g` display table with gdp on person 
*  `-c` display table with co2_emission on person
*  `-s` choose the start considerable year
*  `-e` choose the end considerable year

Example of running:
```sh
 $ python3 project.py -c -s 1999 -e 2001 population.csv gdp.csv co2_emission.csv
 $ python3 project.py -g -s 1999 -e 2001 population.csv gdp.csv co2_emission.csv
```

````python
#importing required libraries
import os

try:
    import numpy as np
except ImportError:
    print("pandas is not installed. Installing numpy...")
    os.system("sudo apt install python3-numpy")

try:
    import pandas as pd
except ImportError:
    print("pandas is not installed. Installing pandas...")
    os.system("sudo apt install python3-pandas")
````

````python
#importing data:
#table with population
population = pd.read_csv("population.csv", skiprows=4)
population = population.drop("Unnamed: 66", axis=1) #column with spaces
population
````

````python
#table with co2 emissions
co2_emissions = pd.read_csv("co2_emission.csv", skiprows=0)
co2_emissions
````

````python
#table with gdp
gdp = pd.read_csv("gdp.csv", skiprows=4)
gdp = gdp.drop("Unnamed: 66", axis=1)
gdp
````

````python
# 1. TASK: displaying the table of 5 countries with greatest value of gdp / population for each year
import task1
start_year = 1990
end_year = 2000

for i in range(start_year,end_year):
    y = str(i)
    cnl = population["Country Name"].to_numpy()
    pl = population[y].to_numpy()
    gdpl = gdp[y].to_numpy()

    tab = task1.find_5_countries_gdp(cnl, pl, gdpl, y)
    print(tab)
````

````python
# 2. TASK:
import task2
tab2 = task2.co2_emissions_per_person(population, co2_emissions, start_year, end_year)
tab2 #table of sorted co2_emissions with respect to years and countries
````

````python
#TASK2.1
#find table of 5 countries with greatest co2/emissions on person for each year
tab2.groupby(level='Year')['Total'].nlargest(5)
````
