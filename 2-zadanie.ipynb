```python
import random 
lista2 = [[] for i in range(11)] #witch bootles person 'i' have to drink,10th - who is infected
special = random.randint(0, 999)

#to many infected: 
to_many = [511, 767, 895, 959, 991]

def change(special, to_many):
    for i in range(5):
        if(special == to_many[i]):
            special = 1000 + i
    return special

def rchange(special, to_many):
    if special >= 1000:
        special = to_many[special-1000]
    return special
    
special = change(special, to_many)
    

def people(lista, special):
    for i in range(1005):
        if i != 511 and i != 767 and i != 895 and i != 959 and i != 991:
            for k in range(10):
                i2 = i//2**k
                if i2 % 2 == 1:
                    lista[k].append(i)
                    if i == special:
                        lista[10].append(k)


def print_b(lista): #1dim lis
    for el in lista:
        if(el < 1000):
            print(el+1, end=" ")
        else:
            print(to_many[el-1000]+1, end=" ")
    print("\n")
        
#answer:
people(lista2, special)

special = rchange(special, to_many)
print("Special bottle number is ", special+1, "\n")


print("Total number of brave adventurers that will drink from special bottle is", len(lista2[10]), "\n")
print_b(lista2[10])

for i in range(10):
    print("Person", i+1, "will drink from bottles")
    print_b(lista2[i])
    
```
